```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive')

# Load Dataset (Update path as needed)
file_path = '/content/drive/My Drive/car_data_250.csv'
df = pd.read_csv(file_path)

# Standardize column names
df.columns = df.columns.str.strip()

# Display first few rows
display(df.head())

# Histogram Analysis
plt.figure(figsize=(10,6))
sns.histplot(df['Price'], bins=30, kde=True)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Car Prices')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure column names are stripped of spaces
df.columns = df.columns.str.strip()

# Check if 'Fuel_Type' exists before plotting
if 'Fuel_Type' in df.columns:
    # Scatterplot of Horsepower vs. Mileage
    plt.figure(figsize=(10,6))
    sns.scatterplot(x=df['Horsepower'], y=df['Mileage'], hue=df['Fuel_Type'])
    plt.xlabel('Horsepower')
    plt.ylabel('Mileage')
    plt.title('Horsepower vs. Mileage')
    plt.legend(title='Fuel Type')
    sns.regplot(x=df['Horsepower'], y=df['Mileage'], scatter=False, color='black')
    plt.show()
else:
    print("Column 'Fuel_Type' not found in dataset.")

# Bar Chart: Number of Cars per Brand
if 'Brand' in df.columns:
    plt.figure(figsize=(12,6))
    df['Brand'].value_counts().plot(kind='bar', color='skyblue')
    plt.xlabel('Brand')
    plt.ylabel('Number of Cars')
    plt.title('Number of Cars per Brand')
    plt.xticks(rotation=45)
    plt.show()
else:
    print("Column 'Brand' not found in dataset.")

# Grouped Bar Chart: Fuel Type vs Manufacturing Year
if 'Manufacturing_Year' in df.columns and 'Fuel_Type' in df.columns:
    plt.figure(figsize=(12,6))
    sns.countplot(data=df, x='Manufacturing_Year', hue='Fuel_Type')
    plt.xlabel('Manufacturing Year')
    plt.ylabel('Count')
    plt.title('Fuel Type Distribution Over the Years')
    plt.xticks(rotation=45)
    plt.legend(title='Fuel Type')
    plt.show()
else:
    print("Columns 'Manufacturing_Year' or 'Fuel_Type' not found in dataset.")

# Insights
if 'Price' in df.columns:
    print("1. Most common price range:")
    print(df['Price'].describe())

if 'Horsepower' in df.columns and 'Mileage' in df.columns:
    print("\n2. Correlation between Horsepower and Mileage:")
    print(df[['Horsepower', 'Mileage']].corr())

if 'Brand' in df.columns:
    print("\n3. Most popular car brand:")
    print(df['Brand'].value_counts().idxmax())

if 'Manufacturing_Year' in df.columns and 'Fuel_Type' in df.columns:
    print("\n4. Fuel type trends over the years:")
    print(df.groupby('Manufacturing_Year')['Fuel_Type'].value_counts())
```
